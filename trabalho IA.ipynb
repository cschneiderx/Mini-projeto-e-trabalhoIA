Limpando as colunas agent,company

```python
dataset['company'] = dataset['company'].fillna(0)
dataset['agent'] = dataset['agent'].fillna(0)
```

Após retirar valores nulos

```python
print(dataset.isna().sum())
```

```python
linhas_com_nulos_country = dataset[dataset["country"].isnull()]
```

```python
paises_existentes = dataset['country'].dropna().unique()

# Preenche os NaN com um valor aleatório da lista
dataset['country'] = dataset['country'].apply(
    lambda x: random.choice(paises_existentes) if pd.isna(x) else x
)
```

```python
print(dataset.isna().sum())
```

```python
linhas_com_nulos_children = dataset[dataset["children"].isnull()]
print(linhas_com_nulos_children)
```

```python
dataset['children'] = dataset['children'].fillna(dataset['children'].mean())
```

```python
print(dataset.isna().sum())
```

Conectando banco

```python
banco = sqlite3.connect('BancoDeDados.db')
```

```python
dataset.to_sql("dados", banco, if_exists="replace", index=False)
```

```python
consulta = pd.read_sql_query("""
SELECT country, AVG(lead_time) as media_avaliacao
FROM dados
GROUP BY country
ORDER BY media_avaliacao DESC
LIMIT 10;
""", banco)

print(consulta)
```

SQL para testar

```python
taxaCancelamentoPorPais= pd.read_sql_query("""SELECT country,
       COUNT(*) AS total_reservas,
       SUM(is_canceled) AS total_canceladas,
       ROUND(SUM(is_canceled) * 100.0 / COUNT(*), 2) AS taxa_cancelamento
FROM dados
GROUP BY country
HAVING total_reservas > 50
ORDER BY taxa_cancelamento DESC
LIMIT 10;""", banco)

print(taxaCancelamentoPorPais)
```

```python
ocupacaoMediaHotel = pd.read_sql_query("""
SELECT hotel,
       ROUND(AVG(stays_in_week_nights + stays_in_weekend_nights), 2) AS media_noites
FROM dados
GROUP BY hotel
ORDER BY media_noites DESC;
""", banco)

print(ocupacaoMediaHotel)
```

```python
receitaPorCanal = pd.read_sql_query("""
SELECT distribution_channel,
       ROUND(AVG(adr), 2) AS media_diaria,
       COUNT(*) AS total_reservas
FROM dados
GROUP BY distribution_channel
ORDER BY media_diaria DESC;
""", banco)

print(receitaPorCanal)
```

```python
precoMedioPorMes = pd.read_sql_query("""
SELECT arrival_date_month,
       ROUND(AVG(adr), 2) AS media_preco
FROM dados
GROUP BY arrival_date_month
ORDER BY media_preco DESC;
""", banco)

print(precoMedioPorMes)
```

```python
mediaCriancasPorCliente = pd.read_sql_query("""
SELECT customer_type,
       ROUND(AVG(children), 2) AS media_criancas
FROM dados
GROUP BY customer_type
ORDER BY media_criancas DESC;
""", banco)

print(mediaCriancasPorCliente)
```

```python
cancelamentoPorRepeticao = pd.read_sql_query("""
SELECT is_repeated_guest,
       ROUND(AVG(is_canceled) * 100, 2) AS taxa_cancelamento
FROM dados
GROUP BY is_repeated_guest;
""", banco)

print(cancelamentoPorRepeticao)
```

```python
precoPorEstacionamento = pd.read_sql_query("""
SELECT required_car_parking_spaces,
       COUNT(*) AS total,
       ROUND(AVG(adr), 2) AS media_preco
FROM dados
GROUP BY required_car_parking_spaces
ORDER BY required_car_parking_spaces;
""", banco)

print(precoPorEstacionamento)
```

```python
comparacaoQuartos = pd.read_sql_query("""
SELECT reserved_room_type,
       assigned_room_type,
       COUNT(*) AS total
FROM dados
WHERE reserved_room_type != assigned_room_type
GROUP BY reserved_room_type, assigned_room_type
ORDER BY total DESC
LIMIT 10;
""", banco)

print(comparacaoQuartos)
```

```python
receitaPorPais = pd.read_sql_query("""
SELECT country,
       ROUND(SUM(adr * (stays_in_week_nights + stays_in_weekend_nights)), 2) AS receita_total
FROM dados
GROUP BY country
HAVING receita_total > 0
ORDER BY receita_total DESC
LIMIT 10;
""", banco)

print(receitaPorPais)
```

```python
rankingPedidosEspeciais = pd.read_sql_query("""
SELECT country,
       COUNT(*) AS total_reservas,
       SUM(total_of_special_requests) AS total_pedidos,
       RANK() OVER (ORDER BY SUM(total_of_special_requests) DESC) AS posicao
FROM dados
GROUP BY country
LIMIT 10;
""", banco)

print(rankingPedidosEspeciais)
```
