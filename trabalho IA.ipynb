```python
import pandas as pd
import random
import sqlite3

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', False)
dataset = pd.read_csv('hotel_bookings.csv', sep=',', index_col= None)
print('A')
```

Limpando as colunas agent,company

```python
dataset['company'] = dataset['company'].fillna(0)
dataset['agent'] = dataset['agent'].fillna(0)
```

Após retirar valores nulos

```python
print(dataset.isna().sum())
```

```python
linhas_com_nulos_country = dataset[dataset["country"].isnull()]
```

```python
paises_existentes = dataset['country'].dropna().unique()

# Preenche os NaN com um valor aleatório da lista
dataset['country'] = dataset['country'].apply(
    lambda x: random.choice(paises_existentes) if pd.isna(x) else x
)
```

```python
print(dataset.isna().sum())
```

```python
linhas_com_nulos_children = dataset[dataset["children"].isnull()]
print(linhas_com_nulos_children)
```

```python
dataset['children'] = dataset['children'].fillna(dataset['children'].mean())
```

```python
print(dataset.isna().sum())
```

Conectando banco

```python
banco = sqlite3.connect('BancoDeDados.db')
```

```python
dataset.to_sql("dados", banco, if_exists="replace", index=False)
```

```python
consulta = pd.read_sql_query("""
SELECT country, AVG(rating) as media_avaliacao
FROM dados
GROUP BY country
ORDER BY media_avaliacao DESC
LIMIT 10;
""", banco)

print(consulta)
```
